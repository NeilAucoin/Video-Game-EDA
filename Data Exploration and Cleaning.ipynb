```python
# Clean up columns with messy formatting

# Starting index
start_index = 0
    # Re-ran after adding more data

# platforms
# Replace NaN values with an empty list
df.loc[start_index:, 'platforms'] = df.loc[start_index:, 'platforms'].fillna('[]')
# Convert strings into lists of dictionaries
df.loc[start_index:, 'platforms'] = df.loc[start_index:, 'platforms'].apply(lambda x: [platform['platform']['name'] for platform in ast.literal_eval(x)])
    
# parent platforms
# Replace NaN values with an empty list
df.loc[start_index:, 'parent_platforms'] = df.loc[start_index:, 'parent_platforms'].fillna('[]')
# Convert strings into lists of dictionaries
df.loc[start_index:, 'parent_platforms'] = df.loc[start_index:, 'parent_platforms'].apply(lambda x: [platform['platform']['name'] for platform in ast.literal_eval(x)])
    
# genres
# Replace NaN values with an empty list
df.loc[start_index:, 'genres'] = df.loc[start_index:, 'genres'].fillna('[]')
# Convert strings into lists of dictionaries
df.loc[start_index:, 'genres'] = df.loc[start_index:, 'genres'].apply(lambda x: [genre['name'] for genre in ast.literal_eval(x)])
    
# stores
# Replace NaN values with an empty list
df.loc[start_index:, 'stores'] = df.loc[start_index:, 'stores'].fillna('[]')
# Convert strings into lists of dictionaries
df.loc[start_index:, 'stores'] = df.loc[start_index:, 'stores'].apply(lambda x: [store['store']['name'] for store in ast.literal_eval(x)])    

# tags
# Replace NaN values with an empty list
df.loc[start_index:, 'tags'] = df.loc[start_index:, 'tags'].fillna('[]')
# Convert strings into lists of dictionaries
df.loc[start_index:, 'tags'] = df.loc[start_index:, 'tags'].apply(lambda x: [tag['name'] for tag in ast.literal_eval(x)])

# esrb ratings
# Replace NaN values with an empty dictionary
df.loc[start_index:, 'esrb_rating'] = df.loc[start_index:, 'esrb_rating'].fillna('{}')
# Convert strings into lists of dictionaries
df.loc[start_index:, 'esrb_rating'] = df.loc[start_index:, 'esrb_rating'].apply(lambda x: ast.literal_eval(x).get('name', ''))

# Show the first few rows of the DataFrame
df.head(3)
```

```python
# Identify Platforms, Genres, Stores, Tags, and ESRB Ratings

# Columns
columns_to_analyze = ['released', 'platforms', 'parent_platforms', 'genres', 'stores', 'tags', 'esrb_rating']

# Dictionary to store values and counts
unique_values_counts = {}

# Function to strip brackets from values
def strip_brackets(value):
    return value.strip("[]'")

# Loop through each column
for column in columns_to_analyze:
    # Combine all values into one string
    if column == 'released':
        combined_values = df[column].apply(lambda x: str(x)[:4]).dropna()
    else:
        combined_values = ', '.join(df[column].dropna()).split(', ')
    # Strip brackets
    stripped_values = [strip_brackets(value) for value in combined_values]
    # Extract values and counts
    unique_values_counts[column] = pd.Series(stripped_values).value_counts().to_dict()

# Print values and counts that appear 5+ times
for column, values_counts in unique_values_counts.items():
    print(f"Unique values and their counts for '{column}':")
    for value, count in values_counts.items():
        if count >= 5:
            print(f"{value}: {count}")
    print()
```

```python
# Function to remove brackets
def remove_brackets(cell):
    if isinstance(cell, str):
        return cell.replace('[', '').replace(']', '')
    else:
        return cell

# Apply function
df = df.applymap(remove_brackets)

# Print results
df.head(3)
```

```python
# Average out one total rating from ratings and metacritic

# Fill null values with 0
df['rating'].fillna(0, inplace=True)
df['metacritic'].fillna(0, inplace=True)

# Function to calculate overall rating
def calculate_overall_rating(row):
    if row['rating'] == 0:
         # Normalize Metacritic score to scale of 1 to 5
        return row['metacritic'] / 20
    elif row['metacritic'] == 0:
        return row['rating']
    else:
        return (row['rating'] + row['metacritic'] / 20) / 2

# Apply function to each row
df['overall_rating'] = df.apply(calculate_overall_rating, axis=1)

# Drop the rating and Metacritic columns
df.drop(['rating', 'metacritic'], axis=1, inplace=True)

# Print results
df.head(3)
```
